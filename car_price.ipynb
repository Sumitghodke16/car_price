```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame (Replace this with your actual DataFrame)
data = {'Fuel_Type': ['Petrol', 'Diesel', 'Petrol', 'CNG', 'Diesel', 'Petrol'],
        'Transmission': ['Manual', 'Automatic', 'Manual', 'Manual', 'Automatic', 'Automatic']}

df = pd.DataFrame(data)

# Count unique values in each column
fuel_counts = df['Fuel_Type'].value_counts()
transmission_counts = df['Transmission'].value_counts()

# Plotting
plt.figure(figsize=(8, 5))

plt.bar(fuel_counts.index, fuel_counts.values, label='Fuel Type', alpha=0.7)
plt.bar(transmission_counts.index, transmission_counts.values, label='Transmission', alpha=0.7)

plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Fuel Type and Transmission Count')
plt.legend()
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv(r"C:\Users\Sumit\Downloads\car_price.csv")

# Group by Brand and Fuel_Type and count the occurrences
brand_fuel_type_count = df.groupby(['Brand', 'Fuel_Type']).size().unstack(fill_value=0)

# Create a bar plot for the data
plt.figure(figsize=(12, 6))
brand_fuel_type_count.plot(kind='bar', stacked=True, colormap='tab20')

# Add titles and labels
plt.title('Car Brands vs Fuel Type Distribution', fontsize=16)
plt.xlabel('Brand', fontsize=12)
plt.ylabel('Count of Cars', fontsize=12)
plt.xticks(rotation=45, ha='right')  # Rotate brand names for readability
plt.legend(title='Fuel Type', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()
```
